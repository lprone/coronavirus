```python
import pandas as pd
```

```python
df = pd.read_json('https://argentina-covid19-data.now.sh/api/v0/daily')
```

```python
df
```

```python
# {'confirmed': 1}

def extractData(data):
    try:
        return int(str(data).replace("{'confirmed': ", "").replace("}", ""))
    except:
        print(data)
        return data;
```

```python
df.columns
```

```python
for column in df.columns:
    df[column] = df[column].apply(lambda x: extractData(x))
```
